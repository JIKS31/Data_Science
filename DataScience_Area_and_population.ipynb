```python
import pandas as pd
cities=pd.read_csv(r"D:\PROJECTS\Datasets_mini_veruthe\california_cities.csv")
```

```python
print(cities.head())
```

```python
latitude,longitude = cities["latd"],cities["longd"]
population,area=cities["population_total"],cities["area_total_km2"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
cities.plot()
plt.show()
```

```python
plt.scatter(longitude, latitude, label=None, c=np.log10(population),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')  # Corrected ylabel
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Create a legend, plot empty lists with the desired size and label
for area_size in [100, 300, 500]:  # Changed loop variable to avoid conflict
    plt.scatter([], [], c='k', alpha=0.3, s=area_size, label=str(area_size) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')

# Title and show plot
plt.title("Area and Population of California Cities")
plt.show()
```
